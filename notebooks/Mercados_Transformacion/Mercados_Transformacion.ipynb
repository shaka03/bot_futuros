**ACÁ SE HACE EL AJUSTE DE LA TRANSFORMACIÓN DE LOS DATASETS DE PRECIOS DE CIERRE DE MTB NTB DTB Y ELN**

```python

# Diccionario de meses
mapa_meses = {
    'F': 'Enero',
    'G': 'Febrero',
    'H': 'Marzo',
    'J': 'Abril',
    'K': 'Mayo',
    'M': 'Junio',
    'N': 'Julio',
    'Q': 'Agosto',
    'U': 'Septiembre',
    'V': 'Octubre',
    'X': 'Noviembre',
    'Z': 'Diciembre'
}
def transformar_dataset(df):
    # Asegurar que Nemo está como fecha
    df['Nemo'] = pd.to_datetime(df['Nemo'], dayfirst=True)

    # Formato largo
    df_long = df.melt(
        id_vars='Nemo',
        var_name='Contrato',
        value_name='Precio'
    )

    # Tipo = primeras 3 letras
    df_long['Tipo'] = df_long['Contrato'].str[:3]

    # Mes = letra 4 convertida
    df_long['Mes'] = df_long['Contrato'].str[3].map(mapa_meses)

    # Año = posiciones 5-6 convertidas a 20xx
    df_long['Año'] = df_long['Contrato'].str[4:6].astype(int).apply(lambda x: 2000 + x)

    # Fecha = Nemo original
    df_long['Fecha'] = df_long['Nemo']

    # Reordenar columnas
    df_final = df_long[['Contrato', 'Tipo', 'Mes', 'Año', 'Fecha', 'Precio']]

    return df_final


```

**DATASETS ORGANIZADOS**

```python
MTBF = transformar_dataset(cierre_mtb)
NTBF = transformar_dataset(cierre_ntb)
DTBF = transformar_dataset(cierre_dtb)
ELMF = transformar_dataset(cierre_dtb)

MTBF= MTBF.rename(columns={
    'Contrato': 'Nemotecnico'
})

NTBF= NTBF.rename(columns={
    'Contrato': 'Nemotecnico'
})

DTBF= DTBF.rename(columns={
    'Contrato': 'Nemotecnico'
})
ELMF= ELMF.rename(columns={
    'Contrato': 'Nemotecnico'
})
```

```python
neg_electronica = neg_electronica.rename(columns={
    'MEJOR PRECIO DE COMPRA (BID)': 'BID',
    'CANTIDAD DE CONTRATOS CON EL MEJOR PRECIO DE COMPRA': 'CABID',
    'MEJOR PRECIO DE VENTA (OFFER)': 'OFFER',
    'CANTIDAD DE CONTRATOS CON EL MEJOR PRECIO DE VENTA': 'CAOFFER'
})
```

```python
# Mapa de meses
mapa_meses = {
    'F': 'Enero',
    'G': 'Febrero',
    'H': 'Marzo',
    'J': 'Abril',
    'K': 'Mayo',
    'M': 'Junio',
    'N': 'Julio',
    'Q': 'Agosto',
    'U': 'Septiembre',
    'V': 'Octubre',
    'X': 'Noviembre',
    'Z': 'Diciembre'
}

def transformar_neg(df):
    # Asegurar que FECHA esté como datetime
    df['FECHA'] = pd.to_datetime(df['FECHA'], dayfirst=True)

    # Renombrar CONTRATO → Contrato para consistencia
    df = df.rename(columns={'CONTRATO': 'Contrato'})

    # Tipo = primeras 3 letras
    df['Tipo'] = df['Contrato'].str[:3]

    # Mes = 4ta letra convertida
    df['Mes'] = df['Contrato'].str[3].map(mapa_meses)

    # Año = posiciones 5-6 → 20xx
    df['Año'] = df['Contrato'].str[4:6].astype(int).apply(lambda x: 2000 + x)

    return df
```

## Variables de los df ELMF-DTBF-MTBF-NTBF

| **Columna**    | **Descripción** |

### Variables de df resumen

Este dataset consolida la cantidad total de energía negociada en el mercado de futuros de energía, agrupada por mes y año. Cada fila representa el volumen total transado en un mes específico.

| Columna | Descripción |
|--------|-------------|
| **Mes** | Fecha en formato `YYYY-MM-01` que representa el mes de negociación. Esta fecha funciona como marcador temporal del mes consolidado. |
| **kWh** | Cantidad total de energía negociada durante el mes, expresada en kilovatios-hora (kWh). |
| **GWh** | Conversión del valor anterior a gigavatios-hora (GWh). Se obtiene dividiendo `kWh / 1,000,000`. |
| **Año** | Año correspondiente al período de negociación. Extraído directamente de la columna *Mes*. |
| **Mes (nombre)** | Nombre del mes (Enero, Febrero, Marzo…), generado para facilitar análisis y visualizaciones. |
