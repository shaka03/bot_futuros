```python
from pydataxm.pydatasimem import CatalogSIMEM, ReadSIMEM
import datetime as dt
import pandas as pd
import calendar
```

```python
# Buscar dataset
cat = CatalogSIMEM("Datasets")
df_cat = cat.get_data()

listado = set(df_cat["nombreConjuntoDatos"])

for i in listado:
    if "bilateral" in i.lower():
        print(i)
```

```python
nombre = "Precios y estadísticas mensuales de contratos entre agentes"
df_cat[df_cat["nombreConjuntoDatos"] == nombre]
```

```python
# Descargar
nombre = "Precios horarios y estadísticas de los contratos entre agentes"
id_dataset = df_cat[df_cat["nombreConjuntoDatos"] == nombre]["idDataset"].values[0]
fecha_inicio_str = "2026-02-01"
fecha_inicio = dt.datetime.strptime(fecha_inicio_str, "%Y-%m-%d")
fecha_fin_str = "2026-02-28"
fecha_fin = dt.datetime.strptime(fecha_fin_str, "%Y-%m-%d")
list_df = []
keep = True

print(f"Iniciando descarga del dataset '{nombre}', id {id_dataset}")
while keep:
        fecha_part = dt.datetime(fecha_inicio.year, 12, 31)
        fecha_part_str = dt.datetime.strftime(fecha_part, "%Y-%m-%d")
        print(f"{fecha_inicio_str} a {fecha_part_str}")
        if fecha_part >= fecha_fin:
            reader = ReadSIMEM(id_dataset, fecha_inicio_str, fecha_fin_str)
            keep = False
        else:
            reader = ReadSIMEM(id_dataset, fecha_inicio_str, fecha_part_str)
        
        df = reader.main(filter=False)
        list_df.append(df)
        del df

        fecha_inicio = dt.datetime(fecha_inicio.year + 1, 1, 1)
        fecha_inicio_str = dt.datetime.strftime(fecha_inicio, "%Y-%m-%d")

df = pd.concat(list_df)
df.head()
```
