### **Análisis Univariado**

#### **Futuros**

```python
# Carga de datos
df_futuros = pd.read_csv(SILVER_DATA_PATH + dict_files["futuros"], parse_dates=["Fecha", "FechaVencimientoContrato"])

# Filtrar por rango de fechas
mask = (df_futuros["Fecha"] >= start_date) & (df_futuros["Fecha"] <= end_date)
df_futuros = df_futuros.loc[mask]

# Forma del DataFrame
print(f"El DataFrame de futuros tiene {df_futuros.shape[0]} filas y {df_futuros.shape[1]} columnas.")

# Revisar las primeras filas del DataFrame
display(df_futuros.head())
```

```python
# Revisar la estructura del DataFrame
df_futuros.info()
```

```python
# Revisar si hay valores nulos
print("\nValores nulos por columna:")
df_futuros.isnull().sum()
```

```python
# Estadísticas descriptivas, histograma y violin plot para cada tipo de contrato
cols_analyze = ["Fecha", "FechaVencimientoContrato", "Precio"]
for tipo in df_futuros["Tipo"].unique():
    print(f"\nEstadísticas para contratos de tipo: {tipo}")
    temp_df = df_futuros[df_futuros["Tipo"] == tipo]
    display(temp_df[cols_analyze].describe())

    # 1. Crear la estructura de la figura (1 fila, 2 columnas)
    datos_filtrados = temp_df["Precio"] # Asegurarse de no incluir NaN en los gráficos
    fig = make_subplots(
        rows=1, cols=1,
        subplot_titles=(f"Histograma de Precios - {tipo}")
    )

    # 2. Agregar el Histograma (Columna 1)
    # Nota: Plotly puro no trae la curva KDE automática como Seaborn, 
    # pero el histograma interactivo es muy potente.
    fig.add_trace(
        go.Histogram(
            x=datos_filtrados,
            name="Histograma",
            marker_color='#1f77b4',  # Un azul estándar
            opacity=0.75
        ),
        row=1, col=1
    )

    # 3. Actualizar el diseño general (Layout)
    fig.update_layout(
        title_text=f"Análisis de Distribución - {tipo}",
        showlegend=False,           # Ocultamos la leyenda para limpiar la vista
        template="plotly_white",    # Fondo blanco estilo Seaborn
        height=500,                 # Altura similar a figsize=(10,4)
        width=500
    )

    # Etiquetas de los ejes
    fig.update_xaxes(title_text="Precio (COP/kWh)", row=1, col=1)
    fig.update_yaxes(title_text="Frecuencia", row=1, col=1)

    fig.show()
```

```python
# Violin plot para cada tipo de contrato en un sólo gráfico
colors_dict = {
    "ELM": '#1f77b4',  # Azul
    "MTB": '#ff7f0e',  # Naranja
    "DTB": '#2ca02c',  # Verde
    "NTB": '#d62728',  # Rojo
}
fig = go.Figure()
for tipo in df_futuros["Tipo"].unique():
    datos_filtrados = df_futuros[df_futuros["Tipo"] == tipo]["Precio"].dropna()
    fig.add_trace(
        go.Violin(
            y=datos_filtrados,
            name=tipo,
            box_visible=True,  # Mostrar caja dentro del violín
            meanline_visible=True,  # Mostrar línea de la media
            line_color='black',  # Color de la línea del violín
            fillcolor=colors_dict[tipo],  # Color de relleno según tipo
            opacity=0.75
        )
    )
fig.update_layout(
    title_text="Violin Plot de Precios por Tipo de Contrato",
    template="plotly_white",
    height=500,
    width=1000,
    yaxis_title="Precio (COP/kWh)"
)

fig.show()
```

```python
# Estadísticas descriptivas, histograma y violin plot para generacion
print(f"\nEstadísticas para los datos de generacion:")
display(df_generacion.describe())

cols_generacion = [x for x in df_generacion.columns if x not in ["Fecha"]]
for col in cols_generacion:
    # 1. Crear la estructura de la figura (1 fila, 2 columnas)
    datos_filtrados = df_generacion[col].dropna() # Asegurarse de no incluir NaN en los gráficos
    fig = make_subplots(
        rows=1, cols=1,
        subplot_titles=(f"Histograma de {col}", f"Violin Plot de {col}")
    )

    # 2. Agregar el Histograma (Columna 1)
    # Nota: Plotly puro no trae la curva KDE automática como Seaborn, 
    # pero el histograma interactivo es muy potente.
    fig.add_trace(
        go.Histogram(
            x=datos_filtrados,
            name="Histograma",
            marker_color='#1f77b4',  # Un azul estándar
            opacity=0.75
        ),
        row=1, col=1
    )

    # 4. Actualizar el diseño general (Layout)
    fig.update_layout(
        title_text=f"Análisis de Distribución - {col}",
        showlegend=False,           # Ocultamos la leyenda para limpiar la vista
        template="plotly_white",    # Fondo blanco estilo Seaborn
        height=500,                 # Altura similar a figsize=(10,4)
        width=500
    )

    # Etiquetas de los ejes
    fig.update_xaxes(title_text="Generación (kWh)", row=1, col=1)
    fig.update_yaxes(title_text="Frecuencia", row=1, col=1)

    fig.show()
```

```python
# Violin plot para cada tipo de generación en un sólo gráfico
colors_dict = {
    "Generacion_Cogenerador_kWh": '#1f77b4',  # Azul
    "Generacion_Eolica_kWh": '#ff7f0e',  # Naranja
    "Generacion_Hidraulica_kWh": '#2ca02c',  # Verde
    "Generacion_Solar_kWh": '#d62728',  # Rojo
    "Generacion_Termica_kWh": '#9467bd'  # Morado para la generación térmica
}

cols_generacion = [x for x in colors_dict.keys()]
fig = go.Figure()
for dem in cols_generacion:
    datos_filtrados = df_generacion[dem].dropna()
    fig.add_trace(
        go.Violin(
            y=datos_filtrados,
            name=dem,
            box_visible=True,  # Mostrar caja dentro del violín
            meanline_visible=True,  # Mostrar línea de la media
            line_color='black',  # Color de la línea del violín
            fillcolor=colors_dict[dem],  # Color de relleno según tipo
            opacity=0.75
        )
    )
fig.update_layout(
    title_text=f"Violin Plot de Generación por Tipo",
    template="plotly_white",
    height=500,
    width=1000,
    yaxis_title="Generación (kWh)"
)

fig.show()
```

```python
# % Generación por tipo por año y mes
df_generacion_group = df_generacion.copy()

df_generacion_group["Año"] = df_generacion_group["Fecha"].dt.year

cols_generacion = [x for x in df_generacion_group.columns if x not in ["Fecha", "Año", "Generacion_Total_kWh"]]
df_generacion_group2 = df_generacion_group.groupby(["Año"])[cols_generacion].sum().reset_index()
for col in df_generacion_group2.columns:
    if col in cols_generacion:
        df_generacion_group2[f"%_{col}"] = ((df_generacion_group2[col] / df_generacion_group2[cols_generacion].sum(axis=1)) * 100).round(2)
        df_generacion_group[f"%_{col}"] = ((df_generacion_group[col] / df_generacion_group[cols_generacion].sum(axis=1)) * 100).round(2)

cols_porcentaje = [x for x in df_generacion_group2.columns if x.startswith("%_")]

print("\nPorcentaje de generación por tipo por año:")
display(df_generacion_group2[["Año"] + cols_porcentaje])

# Gráfico de líneas para el porcentaje de generación por tipo a lo largo del tiempo
colors_dict = {
    "%_Generacion_Cogenerador_kWh": '#1f77b4',  # Azul
    "%_Generacion_Eolica_kWh": '#ff7f0e',  # Naranja
    "%_Generacion_Hidraulica_kWh": '#2ca02c',  # Verde
    "%_Generacion_Solar_kWh": '#d62728',  # Rojo
    "%_Generacion_Termica_kWh": '#9467bd'  # Morado para la generación térmica
}

fig = go.Figure()
for col in cols_porcentaje:
    fig.add_trace(
        go.Scatter(
            x=df_generacion_group["Fecha"],
            y=df_generacion_group[col],
            mode='lines+markers',
            name=col.replace("%_", ""),
            line=dict(color=colors_dict[col], width=2)
        )
    )
fig.update_layout(
    title="Porcentaje de Generación por Tipo a lo largo del Tiempo",
    xaxis_title="Fecha",
    yaxis_title="Porcentaje (%)",
    template="plotly_white",
    height=500,
    width=1000,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.4,
        xanchor="center",
        x=0.5
    )
)

fig.show()
```

El día 17 de diciembre de 2025 se presenta una posible anomalía en la generación por parte de los Cogeneradores. Consultando, esta posible anomalía se puede deber a:

1. Compensación por baja generación hidráulica: El análisis de los datos muestra que el 17 de diciembre la generación hidráulica cayó de 202 GWh a 197 GWh. Para mantener el equilibrio del sistema, el despacho nacional solicitó un aumento en la entrega de excedentes de fuentes no hidráulicas.

2. Temporada de Zafra (Cosecha): Diciembre es un mes de alta intensidad en la cosecha de caña de azúcar en Colombia. En esta fecha específica, los ingenios alcanzaron su máxima capacidad de entrega de excedentes debido a la abundancia de biomasa (bagazo) disponible y a la necesidad del sistema de suplir la demanda de fin de año.

3. Prueba de Disponibilidad: Técnicamente, este tipo de picos aislados de un solo día también suelen corresponder a "Pruebas de Disponibilidad" exigidas por XM. Los cogeneradores deben demostrar periódicamente que pueden alcanzar su capacidad máxima declarada para asegurar su remuneración por Cargo por Confiabilidad.
