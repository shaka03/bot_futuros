```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

### **Parámetros**

```python
os.chdir("/Users/juanchacon/Library/Mobile Documents/com~apple~CloudDocs/GitHub/bot_futuros")

SILVER_DATA_PATH = os.getcwd() + "/data/silver/"

dict_files = {
    "futuros": "precios_FUTUROS.csv",
    "precios": "datos_PRECIOS.csv",
    "precios_ponderados": "datos_PRECIOS_PONDERADOS.csv",
    "demanda": "datos_DEMANDA.csv",
    "demanda_comprador": "datos_DEMANDA_COMPRADOR.csv",
    "aportes_hidricos": "datos_APORTES_HIDRICOS.csv",
    "niveles_embalse": "datos_NIVELES_EMBALSE.csv",
    "generacion": "datos_GENERACION_REAL.csv",
    "disponibilidad": "datos_DISPONIBILIDAD_REAL.csv",
}

start_date = "2024-01-01"
end_date = "2026-01-31"
```

### **EDA**

#### **Futuros**

```python
# Carga de datos
df_futuros = pd.read_csv(SILVER_DATA_PATH + dict_files["futuros"], parse_dates=["Fecha", "FechaVencimientoContrato"])

# Filtrar por rango de fechas
mask = (df_futuros["Fecha"] >= start_date) & (df_futuros["Fecha"] <= end_date)
df_futuros = df_futuros.loc[mask]

# Forma del DataFrame
print(f"El DataFrame de futuros tiene {df_futuros.shape[0]} filas y {df_futuros.shape[1]} columnas.")

# Revisar las primeras filas del DataFrame
display(df_futuros.head())
```

```python
# Revisar la estructura del DataFrame
df_futuros.info()
```

```python
# Estadísticas descriptivas, histograma y violin plot para cada tipo de contrato
cols_analyze = ["Fecha", "FechaVencimientoContrato", "Precio"]
for tipo in df_futuros["Tipo"].unique():
    print(f"\nEstadísticas para contratos de tipo: {tipo}")
    display(df_futuros[df_futuros["Tipo"] == tipo][cols_analyze].describe())

    # Histograma
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df_futuros[df_futuros["Tipo"] == tipo]["Precio"], bins=30, kde=True)
    plt.title(f"Histograma de Precios - {tipo}")
    plt.xlabel("Precio")
    plt.ylabel("Frecuencia")

    # Violin plot
    plt.subplot(1, 2, 2)
    sns.violinplot(x=df_futuros[df_futuros["Tipo"] == tipo]["Precio"])
    plt.title(f"Violin Plot de Precios - {tipo}")
    plt.xlabel("Precio")
    plt.tight_layout()
    plt.show()
```
